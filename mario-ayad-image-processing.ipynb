# Importing Libraries



```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import cv2
import os 
from skimage import exposure
```

# Reading Images Path

```python
img= cv2.imread('../input/cars-image-dataset/Cars Dataset/train/Rolls Royce/1.jpg')
```

# Displaying Original Image (RGB Format)

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.title("Original")
plt.show()
```

# Image in Gray Scale

```python
img_gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap="gray")
plt.title("Gray scale")
```

# Image in BGR Format

```python
plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.title("BGR Format")
plt.show()
```

# Image in red color

```python
img_red = img.copy()
img_red[:, :, 1] = 0    
img_red[:, :, 2] = 0 
plt.imshow(img_red)
plt.title("Red")
```

# Image in blue color

```python
img_blue = img.copy()
img_blue[:,:,[0,1]]=0 
plt.figure(figsize=(6, 6))
plt.imshow(img_blue)
plt.title("Blue")
```

```python
img_green = img.copy()
img_green[:,:,[0,2]]=0   
plt.figure(figsize=(6, 6))
plt.imshow(img_green)
plt.title("Green")
```

# Edge Detection

```python
edges = cv2.Canny(image=img, threshold1=150, threshold2=265)
plt.imshow(edges)
```

# Image with different contrasts

# High contrast

```python
high = cv2.convertScaleAbs(img, alpha= 1.5, beta=20)
plt.figure(figsize=(6, 6))
plt.imshow(high)
plt.title("High")
```

```python
low = cv2.convertScaleAbs(img, alpha=0.5)
plt.figure(figsize=(6, 6))
plt.imshow(low)
plt.title("Low")
```

# Histograms

# RGB Histogram for original image

```python
red = img_rgb[:, :, 0]
green = img_rgb[:, :, 1] 
blue = img_rgb[:, :, 2]
plt.hist(red.ravel(), bins=256, color="red",stacked= True)
plt.hist(green.ravel(), bins=256, color="green")
plt.hist(blue.ravel(), bins=256, color="blue")
plt.title('RGB Histogram for original image')
plt.show()
```

# Single Channel (Grayscale) Histogram

```python
histogram = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
plt.plot(histogram, color='k')
plt.title('Gray Scale Histogram')
plt.show()
```

# Histogram Equalization

```python
Hist_Eq = exposure.equalize_adapthist(img_rgb)
plt.figure(figsize=(6, 6))
plt.imshow(Hist_Eq)
```

```python
red = Hist_Eq[:, :, 0]
green = Hist_Eq[:, :, 1] 
blue = Hist_Eq[:, :, 2]
plt.hist(red.ravel(), bins=256, color="red",stacked= True)
plt.hist(green.ravel(), bins=256, color="green")
plt.hist(blue.ravel(), bins=256, color="blue")
plt.title('Equalization RGB Histogram')
plt.show()
```

# Image Bluring

```python
blur_img = cv2.blur(img,(4,3))
plt.figure(figsize=(6, 6))
plt.imshow(blur_img)
plt.title("Blurred Image")
```

# Gaussian Blurring

```python
blurred_img = cv2.GaussianBlur(img,(5,5),10)
plt.figure(figsize=(6, 6))
plt.imshow(blurred_img)
plt.title("Gaussian Blurred Image")
```

# Adding text on image

```python
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text='Rolls Royce',org=(1,100), fontFace=font,fontScale=1,color=(120,150,900),thickness=3)
plt.figure(figsize=(6, 6))
plt.imshow(img)
```
